```python
drop_cols = [
    'Customer ID', 'Country', 'State', 'City', 'Zip Code',
    'Latitude', 'Longitude',
    'Customer Status', 'Churn Category', 'Churn Reason',
    'Churn Score', 'CLTV', 'Satisfaction Score',
    'Total Revenue', 'Total Refunds',
    'Total Extra Data Charges', 'Total Long Distance Charges',
    'Offer'
]
```

```python
df_clean=df.drop(columns = drop_cols)
```

```python
df_encoded = pd.get_dummies(df_clean, drop_first=True)
```

```python
y = df_clean['Churn']
X = df_clean.drop('Churn', axis=1)
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
X_encoded.shape

```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#model = LogisticRegression(max_iter=1000)
#this is hitting limit on iterations and not converging , possibly due to shear variety in data ranges
#we will be using feature scaling inorder to fix this
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
y_prob = model.predict_proba(X_test_scaled)[:, 1]
```

```python
y_prob
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, pos_label='Yes'))
print("Recall:", recall_score(y_test, y_pred, pos_label='Yes'))
print("F1 Score:", f1_score(y_test, y_pred, pos_label='Yes'))
print("ROC-AUC:", roc_auc_score((y_test == "Yes").astype(int), y_prob))

```

```python
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------
# Confusion Matrix
# ---------------------------
cm = confusion_matrix(y_test, y_pred, labels=["No", "Yes"])

plt.figure()
sns.heatmap(cm, annot=True, fmt="d",
            xticklabels=["No", "Yes"],
            yticklabels=["No", "Yes"])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ---------------------------
# ROC Curve
# ---------------------------
# y_prob should be probability of positive class "Yes"
# (If not already defined, uncomment next line)
# y_prob = model.predict_proba(X_test_scaled)[:, 1]

fpr, tpr, thresholds = roc_curve((y_test == "Yes").astype(int), y_prob)
auc_score = roc_auc_score((y_test == "Yes").astype(int), y_prob)

plt.figure()
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1])
plt.title(f"ROC Curve (AUC = {auc_score:.3f})")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```
